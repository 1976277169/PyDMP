Probabilistic Movement Primitive (ProMP)
====================================

```python
n_steps = 100
n_demos = 10

def generate_demo(n_steps):
    r1 = random.randn() * 0.05
    r2 = random.randn() * 0.1
    time = linspace(0, 1, n_steps)
    X = ndarray((n_steps, 2))
    X[:, 0] = cos((1+r1) * pi*time)
    X[:, 1] = sin((1+r2) * 4*pi*time)-2*linspace(0, 1, n_steps)
    return X

demos = array([generate_demo(n_steps) for _ in xrange(n_demos)])
for demo in demos:
    plot(demo[:, 0], demo[:, 1], color="b")
```

```python
from dmp import ProMP

tau = 1.0
n_steps = 100
w = random.randn(2, 5)
Sigma = 1e-5*eye(4)
n_samples = 10

promp = ProMP()
promp.imitate(demos)

for i in xrange(n_samples):
    Yn = []
    Ydn = []
    for t in xrange(n_steps):
        s = promp.phase(tau, n_steps, t)
        yn, ydn = promp.sample(tau, w, Sigma, s)
        Yn.append(yn)
        Ydn.append(ydn)
    Yn = array(Yn)
    Ydn = array(Ydn)
    plot(Yn[:, 0], Yn[:, 1], "pink")

Y = []
Yd = []
for t in xrange(n_steps):
    s = promp.phase(tau, n_steps, t)
    y, yd = promp.step(tau, w, s)
    Y.append(y)
    Yd.append(yd)
Y = array(Y)
Yd = array(Yd)
plot(Y[:, 0], Y[:, 1], lw=3)

#figure()
#plot(Yd[:, 0])
#plot(Yd[:, 1])
```
