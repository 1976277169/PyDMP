```python
n_steps = 1000
n_demos = 1

def generate_demo(n_steps):
    r1 = random.randn() * 0.1
    r2 = random.randn() * 0.1
    time = linspace(0, 1, n_steps)
    X = ndarray((n_steps, 2))
    X[:, 0] = cos((1+r1) * pi*time)
    X[:, 1] = sin((1+r2) * 4*pi*time)-2*linspace(0, 1, n_steps)
    return X

demos = array([generate_demo(n_steps) for _ in xrange(n_demos)])
for demo in demos:
    plot(demo[:, 0], demo[:, 1], color="b")
```

Dynamical Movement Primitive (DMP)
===================================

DMPs are trajectory representations that

* are goal-directed
* are robust to perturbations and noise (like potential fields)
* are arbitrarily shapeable
* can be scaled and translated arbitrarily, even online
* are parametrized policy representations that can be used for reinforcement learning
* can be used for imitation learning

They have drawbacks:

* an inverse (velocity) kinematics is required to execute task space trajectories
* they are usually not able to learn from multiple demonstrations

There are different variants of DMPs in the literature [1, 2, 3].

[1] Ijspeert, Auke Jan; Nakanishi, Jun; Hoffmann, Heiko; Pastor, Peter; Schaal, Stefan: Dynamical movement primitives: learning attractor models for motor behaviors, Neural computation 25 (2), pp. 328-373, 2013.

[2] Pastor, Peter; Hoffmann, Heiko; Asfour, Tamim; Schaal, Stefan: Learning and generalization of motor skills by learning from demonstration, Proceedings of the IEEE International Conference on Robotics and Automation, pp. 763-768, 2009.

[3] Muelling, Katharina; Kober, Jens; Kroemer, Oliver; Peters, Jan: Learning to Select and Generalize Striking Movements in Robot Table Tennis, International Journal of Robotics Research 32 (3), 2013.

```python
from dmp import DMP

random.seed(5)


tau = 1.0
n_steps = 1000
n_features = 50
o = array([0.8, 0.0])
dt = tau / n_steps

dmp = DMP()
D = dmp.design_matrix(n_features, tau, demos[0])
f, x0, g = dmp.forces(tau, demos[0])
w = linalg.pinv(D).dot(f).T

x_range = (-0.2, 1.2)
y_range = (-0.2, 1.2)
n_tics = 10

def trajectory(dmp, shape, avoidance):
    T = [x0.copy()]
    To = [x0.copy()]
    F = []
    x = x0.copy()
    xd = zeros_like(x, dtype=numpy.float64)
    xo = x0.copy()
    xod = zeros_like(x, dtype=float64)
    for t in xrange(n_steps):
        s = dmp.phase(tau, n_steps, t)
        C = dmp.obstacle(o, x, xd) if avoidance else 0.0
        f = dmp.forcing_term(x0, g, tau, w, s, x) if shape else 0.0
        xd += dt * (dmp.spring_damper(x0, g, tau, s, x, xd) + f + C)
        x += dt * xd
        T.append(x.copy())
    return array(T)

def potential_field(dmp, t, v):
    X, Y = meshgrid(linspace(x_range[0], x_range[1], n_tics),
                    linspace(y_range[0], y_range[1], n_tics))
    pos = array((X, Y)).transpose((1, 2, 0))
    vel = ones_like(pos)
    for i in xrange(pos.shape[0]):
        for j in xrange(pos.shape[1]):
            vel[i, j] = v
    sd = dmp.spring_damper(x0, g, tau, dmp.phase(tau, n_steps, t), pos, vel)
    f = dmp.forcing_term(x0, g, tau, w, dmp.phase(tau, n_steps, t), pos)
    C = dmp.obstacle(o, pos, vel)
    acc = sd + f + C
    return X, Y, sd, f, C, acc


G = trajectory(dmp, shape=False, avoidance=False)
T = trajectory(dmp, shape=True, avoidance=False)
O = trajectory(dmp, shape=True, avoidance=True)

fig = figure(figsize=(5, 5))

plot(G[:, 0], G[:, 1], lw=3, color="g")
plot(T[:, 0], T[:, 1], lw=3, color="r")
plot(O[:, 0], O[:, 1], lw=3, color="black")

plot(x0[0], x0[1], "o", color="b", markersize=10)
plot(g[0], g[1], "o", color="g", markersize=10)
plot(o[0], o[1], "o", color="y", markersize=10)
```

```python
def potential_trajectory(dmp, max_t, n_steps, shape, avoidance):
    T = [x0.copy()]
    To = [x0.copy()]
    F = []
    x = x0.copy()
    xd = zeros_like(x, dtype=numpy.float64)
    xo = x0.copy()
    xod = zeros_like(x, dtype=float64)
    for t in xrange(max_t):
        s = dmp.phase(tau, n_steps, t)
        C = dmp.obstacle(o, x, xd) if avoidance else 0.0
        f = dmp.forcing_term(x0, g, tau, w, s, x) if shape else 0.0
        xd += dt * (dmp.spring_damper(x0, g, tau, s, x, xd) + f + C)
        x += dt * xd
        T.append(x.copy())
    X, Y, sd, f, C, acc = potential_field(dmp, max_t-1, xd)
    if not avoidance:
        acc -= C
    return array(T), X, Y, sd, f, C, acc

fig = figure(figsize=(15, 15))
n_gap = n_steps / 10
for t in list(range(1, n_steps, n_gap))[:-1]:
    subplot(3, 3, 1+t/n_gap)

    T, X, Y, sd, f, C, acc = potential_trajectory(dmp, t, n_steps, shape=True, avoidance=True)

    plot(T[:, 0], T[:, 1], lw=3, color="black")

    quiver_scale = 2e3
    quiver(X, Y, sd[:, :, 0], sd[:, :, 1], scale=quiver_scale, color="g")
    quiver(X, Y, f[:, :, 0], f[:, :, 1], scale=quiver_scale, color="r")
    quiver(X, Y, C[:, :, 0], C[:, :, 1], scale=quiver_scale, color="y")
    quiver(X, Y, acc[:, :, 0], acc[:, :, 1], scale=quiver_scale, color="black")

    plot(x0[0], x0[1], "o", color="b", markersize=15)
    plot(g[0], g[1], "o", color="g", markersize=15)
    plot(o[0], o[1], "o", color="y", markersize=15)
```
